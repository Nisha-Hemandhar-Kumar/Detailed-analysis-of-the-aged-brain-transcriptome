### IsoformSwitchAnalyzeR measures isoform usage via isoform fraction (IF) values which quantifies the fraction of the parent gene expression originating from a specific isoform (calculated as isoform_exp / gene_exp). 

### The difference in isoform usage is quantified as the difference in isoform fraction (dIF) 

### calculated as IF2 - IF1

### These dIF are used to measure the effect size (like fold changes are in gene/isoform expression analysis).

```r
subsetaSwitchList <- subsetSwitchAnalyzeRlist(
    switchAnalyzeRlist = aSwitchList,
    subset = abs(aSwitchList$isoformFeatures$dIF) > 0.05
)

summary(subsetaSwitchList)
```

```r
subsetaSwitchList$isoformCountMatrix
```

```r
write.csv(subsetaSwitchList$isoformFeatures,"subsetaSwitchList_gencode_illumina.csv")
```

```r
preFilterSwitchList <- preFilter(subsetaSwitchList, 
                                 geneExpressionCutoff = 5,
                                 isoformExpressionCutoff = 3,
                                 IFcutoff=0.05, 
                                 removeSingleIsoformGenes = TRUE
                                ) # preFilter to remove lowly expressed features


summary(preFilterSwitchList)
```

```r
preFilterSwitchList$runInfo
```

```r
# Perform DEXSeq test
DEXSeqSwitchListAnalyzed <- isoformSwitchTestDEXSeq(
    switchAnalyzeRlist = preFilterSwitchList,
    reduceToSwitchingGenes=TRUE, 
    onlySigIsoforms = TRUE)
```
